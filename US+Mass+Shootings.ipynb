```python
a = pd.merge(kdf,age,on=['Title','Date'],how = 'left').drop(['Latitude','Longitude'],axis = 1)
a
```

```python
latest = mr1.iloc[:,[0,1,2,4,5,6,7,-7]]
latest = latest.rename(columns = {'Case':'Title'})
latest.Injured = latest.Injured.str.rstrip('+')
latest['Total victims'] = latest['Total victims'].str.rstrip('+')
latest.Injured = latest.Injured.astype('int')
latest['Total victims'] = latest['Total victims'].astype('int')

latest
final_df1 = pd.concat([latest,a])
final_df1.shape
```

```python
final_df1.drop(['S#'],inplace=True,axis = 1)
final_df1
```

```python
final_df1.Race.value_counts(dropna=False)
```

```python
final_df1.Race=final_df1.Race.str.replace('White American or European American/Some other Race','Some other race')
final_df1.Race=final_df1.Race.str.replace('Native American or Alaska Native','Some other race')
final_df1.Race=final_df1.Race.str.replace('Asian American/Some other race','Some other race')
final_df1.Race=final_df1.Race.str.replace('Black American or African American/Unknown','Some other race')
final_df1.Race=final_df1.Race.str.replace('Other','Some other race')
final_df1.Race=final_df1.Race.str.replace('unclear','Unknown')
final_df1.Race=final_df1.Race.str.replace('black','Black')
final_df1.Race=final_df1.Race.str.replace('[W,w]hite','White')
final_df1.Race=final_df1.Race.str.replace('White ','White')
final_df1.Race.fillna('Unknown',inplace=True)


final_df1.Race.value_counts()

```

```python
final_df1.Race=final_df1.Race.str.replace('WhiteAmerican or European American','White')
final_df1.Race=final_df1.Race.str.replace('Black American or African American','Black')
final_df1.Race=final_df1.Race.str.replace('Some other race','Others')
final_df1.Race=final_df1.Race.str.replace('Asian American','Asian')
```

```python
ax = final_df1.Race.value_counts().plot(kind = 'bar',grid = True)
plt.xlabel('')
plt.ylabel('Count')
plt.title("Murderer's Race" )
for p in ax.patches:
    ax.annotate(format(p.get_height()), (p.get_x()+0.15, p.get_height()+1))
plt.show()
plt.savefig('D:\python files\pandas exercise\working file\Race.png',format = 'png')
plt.close()
```
